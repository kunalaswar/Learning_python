```python
#Step:2 Retriving the column that we have to work
#! Adding the Total column  into  it
df["Total"] = df["Telugu"]+df["English"]+df["Hindi"]+df["Maths"]+df["Science"]+df["Social"]
print(df)
```

```python
#! Adding the percentage column into it
df["percentage"] = df["percentage"] = df["Total"]/600*100
print(round(df),2)
```

```python
# Adding the Grade column into it
#* df.loc["Simple test condition",["colname"]]
df.loc[df["percentage"]>=75,["Grade"]]="DISTINCTION"
print(df)
```

```python
df.loc[((df["percentage"]>=65) & (df["percentage"]<74)),["Grade"]]="FIRST"
print(df)
```

```python

#
# df.loc[((df["percentage"]>=60) & (df["percentage"]<65)),["Grade"]]="SECOND"

df.loc[((df["percentage"]>=60) & (df["percentage"]<65)),["Grade"]]="SECOND"
print(df)   

```
