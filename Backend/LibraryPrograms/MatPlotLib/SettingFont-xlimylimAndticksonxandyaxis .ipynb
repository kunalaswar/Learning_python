```python

import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4])
y = x*2
plt.plot(x,y,ls="solid",lw=6,c="red",marker="o",mec="green",mfc="yellow",ms=10)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Line Graph")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4])
y = x*2
plt.plot(x,y,ls="solid",lw=6,c="red",marker="o",mec="green",mfc="yellow",ms=10)
#* Set the font property 
font1 = {"family":"serif","color":"blue","size":16}
font2 = {"family":""}
plt.xlabel("X-Axis",fontdict=font1)
plt.ylabel("Y-Axis",fontdict=font1)
plt.title("Line Graph",fontdict=font1)
plt.show()
```

```python
#changes the font of the title
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4])
y = x*2
plt.plot(x,y,ls ="solid",lw=6,c="red",marker="o",mec="green",mfc="yellow",ms=10)
font1={"family":"serif","color":"blue","size":16}
font2={"family":"times new roman","color":"hotpink","size":25}
plt.xlabel("X-Axis",fontdict=font1)
plt.ylabel("Y-Axis",fontdict=font1)
plt.title("Line Graph",fontdict=font2)
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4])
y = x*2
plt.plot(x,y,ls="solid",lw=6,c="red",marker="o",mec="green",mfc="yellow",ms=10)
font1={"family":"serif","color":"blue","size":16}
font2={"family":"Comic Sans MS","color":"blue","size":16}
font3={"family":"times new roman","color":"red","size":25}
plt.xlabel("X-Axis",fontdict=font1)
plt.ylabel("Y-Axis",fontdict=font2)
plt.title("Line Graph",fontdict=font3)
plt.show()
```
