```python


#! Given Problem
#Number of Students in  Class		Heights
#			20						4.5 5.5 6.5 3.5 5.5
#									3.5 5.5 6.2 3.5 5.5
#									3.5 5.5 4.5 3.5 5.5
#									2.5,3.5,6.0,6.0,5.5
#Draw the bar Plot the above Given Data

lst=[4.5,5.5,6.5, 3.5 ,5.5,3.5 ,5.5 ,6.2,3.5,5.5,3.5,5.5, 4.5,3.5,5.5,2.5,3.5,6.0,6.0,5.5]
d = {}
for val in lst:
    if val not in d:
        d[val]=1
    else:
        d[val] = d[val]+1
           
print(d)           
```

```python
for k,v in d.items():
    print(f"{k} --->{v}")
```

```python
import pandas as pd
df = pd.DataFrame(d,index=[1])   #* In dict we always pass the index when we pass the dict into the DataFrame
print(df)
```

```python
#In a firm of 400 employees, the percentage of monthly salary 
#saved by each employee is given in the following table. Represent it through a bar graph.
#----------------------------------------------------------------
#		Savings (in percentage)	Number of Employees(Frequency)
#----------------------------------------------------------------
#			20					105
#			30					199
#			40					29
#			50					73
#----------------------------------------------------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
d = {"Savings":[20,30,40,50],
     "NumEmp":[105,199,29,73]}
# Convert Dict object into DataFrame object
df = pd.DataFrame(d)
print("------------------------------------") 
print(df)
print("------------------------------------") 
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
d = {"Saving":[20,30,40,50],
      "NumEmp":[105,199,29,73]}
# Convert Dict object into DataFrame object
df = pd.DataFrame(d)
print("-------------------------")
print(df)
print("-------------------------")
plt.bar(df["Saving"],df["NumEmp"],color = ["green","blue"],width=1)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
d = {"Savings":[20,30,40,50],
     "NumEmp":[105,199,29,73]}
df = pd.DataFrame(d)
print("---------------------")
print(df)
print("---------------------")
plt.barh(df["Savings"],df["NumEmp"],color=["green","blue"],height=1)
plt.show()
```
