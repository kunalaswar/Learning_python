```python

#Drawing Sub  Plotting
import matplotlib.pyplot as plt
plt.subplot(1,2,1)
plt.plot([1,2,3,4],[4,3,2,1])
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("xy-plot-1")
#------------------------------------------------
plt.subplot(1,2,2)
plt.plot([1,2,3,4],[1,2,3,4])
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("xy-plot-2")
#------------------------------------------------
plt.suptitle("Main Plot Contain subplot")
plt.show()

```

```python
#Drawing Sub Plotting
import matplotlib.pyplot as plt
plt.subplot(2,1,1)
plt.plot([1,2,3,4],[4,3,2,1])
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("XY Plot-1")
#---------------------------------
plt.subplot(2,1,2)
plt.plot([1,2,3,4],[1,2,3,4])
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("XY Plot-2")
#---------------------------------
plt.suptitle("Main Plot contains subplots ")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#plot1
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("SALES")

#plot2
x = np.array([0,1,2,3])
y = np.array([10,20,30,40])

plt.subplot(1,2,2)
plt.plot(x,y)
plt.title("INCOME")
plt.suptitle("Sales and Income Plot")
plt.show()
```
