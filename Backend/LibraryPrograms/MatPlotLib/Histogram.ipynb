```python
# Implementation of Histogram
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
markvalues= np.random.randint(10,51,35)
print("Marks values ",markvalues)
marksinterval=np.array([10,20,30,40,50])
plt.hist(markvalues,marksinterval)
plt.show()

```

```python
numberoflamps=np.random.randint(300,1001,400)
Lifetime=[300,400,500,600,700,800,900,1000]
print("Number of Lamps")
print(numberoflamps)
print("---------------------------------------")
plt.hist(numberoflamps,Lifetime)
plt.show()
```
