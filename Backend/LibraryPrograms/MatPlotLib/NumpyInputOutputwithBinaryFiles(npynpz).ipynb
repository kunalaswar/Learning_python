```python

#!         Save Multiple NumPy Array File(.npz Files)   # there is the Extra z is there
#* there the z meaning is that multiple file is zipping together
```

```python
l1 = [10,20,30,40]
l2 = ["RS","TR","SR","KV"]
x = zip(l1,l2)
print(x)
print(type(x))
for i in x:
    #print(i)
    print(f"{i[0]} ---> {i[1]}")
print("------------------")    
for k,v in x:
    print(k,"--->",v)    
```

```python
#Save Multiple NumPy Array File(.npz Files)
import numpy as np
# Create two numpy array
array1 = np.array([2, 6, 10])
array2 = np.array([4, 8, 12])
# save the two arrays into a single file
np.savez('D:\\PYTHON FULL STACK\\LibraryPrograms\\MatPlotLib\\file2.npz', k = array1, v = array2)
print("Multiple Ndarray objects Data saved into the Single File (.npz)")
```

```python

#!  Load Multiple NumPy Array File--Reading the Data from npz files

import numpy as np

# load the saved arrays 
load_data = np.load('D:\\PYTHON FULL STACK\\LibraryPrograms\\MatPlotLib\\file2.npz')

# retrieve the arrays using their names
array1 = load_data['k']
array2 = load_data['v']

# display the loaded arrays
print("--------------------------------------")
print(array1)
print("--------------------------------------")
print(array2)
print("--------------------------------------")
```
