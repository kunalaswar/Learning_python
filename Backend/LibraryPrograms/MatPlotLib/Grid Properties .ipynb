```python

#! Setting the grid 
from matplotlib import pyplot as plt
import numpy as np
x = [2,1,3,4]
y = [2,3,4,1]
plt.plot(x,y,marker="o",mec="red",mfc="yellow",c="red",ls="solid",lw=3,ms=10)
plt.show()
```

```python

#! grid line are Line only so we can apply line property here
#! Setting the grid taking both axis grid
from matplotlib import pyplot as plt
import numpy as np
x = [2,1,3,4]
y = [2,3,4,1]
plt.plot(x,y,marker="o",mec="red",mfc="yellow",c="red",ls="solid",lw=3,ms=10)
plt.grid(axis="both",ls="dotted",lw=3,c="green") #* line property
plt.show()
```

```python

#! grid line are Line only so we can apply line property here
#! Setting the grid taking both axis grid
from matplotlib import pyplot as plt
import numpy as np
x = [2,1,3,4]
y = [2,3,4,1]
plt.plot(x,y,marker="o",mec="red",mfc="yellow",c="red",ls="solid",lw=3,ms=10)
plt.grid(axis="both",ls="dotted",lw=3,c="green") #* line property
plt.xlabel("X-Axis",fontdict={"family":"arial","color":"blue","size":15})
plt.ylabel("Y-Axis",fontdict={"family":"arial","color":"blue","size":15})
plt.title("line Plot",fontdict={"family":"Times New Roman","color":"red","size":25})
plt.show()
```

```python
#! grid line are Line only so we can apply line property here
#! Setting the grid taking both axis grid
from matplotlib import pyplot as plt
import numpy as np
x = [2,1,3,4]
y = [2,3,4,1]
plt.xlim(0,6)
plt.ylim(0,6)

plt.plot(x,y,marker="o",mec="red",mfc="yellow",c="red",ls="solid",lw=3,ms=10)
plt.grid(axis="both",ls="dotted",lw=3,c="green") #* line property
plt.xlabel("X-Axis",fontdict={"family":"arial","color":"blue","size":15})
plt.ylabel("Y-Axis",fontdict={"family":"arial","color":"blue","size":15})
plt.title("line Plot",fontdict={"family":"Times New Roman","color":"red","size":25})
plt.show()
```

```python
#! grid line are Line only so we can apply line property here
#! Setting the grid taking both axis grid
from matplotlib import pyplot as plt
import numpy as np
x = [2,1,3,4]
y = [2,3,4,1]
plt.xlim(0,6)
plt.ylim(0,6)
plt.xticks([1,2,3,4,5,6])
plt.yticks([1,2,3,4,5,6])
plt.plot(x,y,marker="o",mec="red",mfc="yellow",c="red",ls="solid",lw=3,ms=10)
plt.grid(axis="both",ls="dotted",lw=3,c="green") #* line property
plt.xlabel("X-Axis",fontdict={"family":"arial","color":"blue","size":15})
plt.ylabel("Y-Axis",fontdict={"family":"arial","color":"blue","size":15})
plt.title("line Plot",fontdict={"family":"Times New Roman","color":"red","size":25})
plt.show()
```

```python
#! grid line are Line only so we can apply line property here
#! Setting the grid taking both axis grid
from matplotlib import pyplot as plt
import numpy as np
x = [2,1,3,4]
y = [2,3,4,1]
z = [3,4,2,1]
plt.xlim(0,6)
plt.ylim(0,6)
plt.xticks([1,2,3,4,5,6])
plt.yticks([1,2,3,4,5,6])
plt.plot(x,y,marker="o",mec="red",mfc="yellow",c="red",ls="solid",lw=3,ms=10)
plt.plot(x,z,marker="X",mec="black",mfc="blue",c="hotpink",ls="solid",lw=4,ms=8)
plt.grid(axis="both",ls="dotted",lw=3,c="green") #* line property
plt.xlabel("X-Axis",fontdict={"family":"arial","color":"blue","size":15})
plt.ylabel("Y-Axis",fontdict={"family":"arial","color":"blue","size":15})
plt.title("line Plot",fontdict={"family":"Times New Roman","color":"red","size":25})
plt.show()
```

```python
#! grid line are Line only so we can apply line property here
#! Setting the grid taking both axis grid
from matplotlib import pyplot as plt
import numpy as np
x = [2,1,3,4]
y = [2,3,4,1]
plt.xlim(0,6)
plt.ylim(0,6)
plt.xticks([1,2,3,4,5,6])
plt.yticks([1,2,3,4,5,6])
plt.plot(x,y,marker="o",mec="red",mfc="yellow",c="red",ls="solid",lw=3,ms=10)
plt.plot(x,[3,4,2,1],marker="X",mec="black",mfc="blue",c="hotpink",ls="solid",lw=4,ms=8)
plt.grid(axis="both",ls="dotted",lw=3,c="green") #* line property
plt.xlabel("X-Axis",fontdict={"family":"arial","color":"blue","size":15})
plt.ylabel("Y-Axis",fontdict={"family":"arial","color":"blue","size":15})
plt.title("line Plot",fontdict={"family":"Times New Roman","color":"red","size":25})
plt.legend(["India","newzland"])
plt.show()
```
