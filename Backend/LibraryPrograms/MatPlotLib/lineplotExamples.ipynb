```python
#combine together
#Draw the Line Graph for all three weeks
import pandas as pd
from matplotlib import pyplot as plt
df=pd.DataFrame({"Days":["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"],
                 "Week1":[27,29,32,33,33,31,29],
                 "Week2":[29,30,31,33,34,34,31],
                 "Week3":[29,28,25,27,31,30,29]})
print("------------------------------------")
print(df)
print("------------------------------------")
plt.plot(df["Days"],df["Week1"],lw=6,c="red",marker="o")
plt.plot(df["Days"],df["Week2"],lw=6,c="green",marker="X")
plt.plot(df["Days"],df["Week3"],lw=6,c="black",marker="*")
plt.xlabel("WEEK DAYS",fontdict={"family":"Arial","color":"blue","size":15})
plt.ylabel("TEMP",fontdict={"family":"Arial","color":"blue","size":15})
plt.title("TEMP in 3 WEEK DAYS",fontdict={"family":"Arial","color":"blue","size":20})
plt.legend(labels=["Week1","Week2","Week3"])
plt.grid(axis="both",ls="dotted",c="green",lw=2)
plt.show()
```

```python
#! Implement the above Problem with Multiple Sub Plots in same axes
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("D:\\PYTHON FULL STACK\\LibraryPrograms\\MatPlotLib\\tempweeks.csv")
print("-------------------------------------")
print(df)
print("-------------------------------------")   
#*Draw a Sub Plot for days VS week1   
plt.subplot(3,1,1)
plt.plot(df["Days"],df["Week1"],lw=6,c="red",marker="o")
plt.title("First Week",fontdict={"family":"Arial","color":"blue","size":14})
#*Draw a Sub Plot for days VS week2
plt.subplot(3,1,2)
plt.plot(df["Days"],df["Week2"],lw=6,c="green",marker="*")
plt.title("Second Week",fontdict={"family":"Arial","color":"blue","size":14})
#*Draw a Sub Plot for days VS week3
plt.subplot(3,1,3)
plt.plot(df["Days"],df["Week3"],lw=6,c="black",marker="X")
plt.title("Third Week",fontdict={"family":"Arial","color":"blue","size":14})
#* set the super title for all 
plt.suptitle("TEMP in 3 WEEK DAYS",fontdict={"family":"Arial","color":"red","fontsize":20})
plt.show()

```

```python

#! Q) Imagine you  work in Retail company and you want to evaluate the monthly Sales of twop different product over the past year
#* year :12 Month 
#* product sales values range in between 50 to 200 in 12 months------
        #* product_sales_1 = np.randint(50,200,12)
#* product sales values ranges in between 60 to 180 in 12 months------
        #* product_sales_2 = np.randint(60,180,12)

#todo - WE Can not create CSV file for this Because Data is not given only random data is given here
import numpy as np
import matplotlib.pyplot as plt
retcom = pd.DataFrame({"Month":["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"],
                          "Sales_Prod_1":np.random.randint(50,200,12),
                          "Sales_Prod_2":np.random.randint(60,180,12)})
print("Retails Company Data ")
print("-----------------------------------")
print(retcom)
print("-----------------------------------")
plt.plot(retcom["Month"],retcom["Sales_Prod_1"],lw=6,c="red",marker="o")
plt.plot(retcom["Month"],retcom["Sales_Prod_2"],lw=6,c="green",marker="X")
plt.xlabel("Months",fontdict={"family":"Arial","color":"blue","size":15})
plt.ylabel("Sales differ",fontdict={"family":"Arial","color":"blue","size":15})  
plt.title("Sales in Months",fontdict={"family":"Arial","color":"blue","size":20})
plt.legend(labels=["Sales_Prod_1","Sales_Prod_2"])
plt.grid(axis="both",ls="dotted",c="green",lw=2)
plt.show()


#todo  - randon generate hot aahe
```
