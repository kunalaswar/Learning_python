```python

#! Drawing Multiple Plot
#How to plot Multiple Plots (Lines) in same axes
from matplotlib import pyplot as plt 
import numpy as np
x = np.arange(1,6)
# y = x**2
plt.plot(x)  #* Matplotlib thinks x is y-values and assigns x=[0,1,2,3,4].
plt.show()
```

```python
from matplotlib import pyplot as pyplot
import numpy as np
x = np.arange(1,6)
y = x**2
plt.plot(x,x)
plt.show()
```

```python
#How to plot Multiple Plots (Lines) in same axes
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(1,6)
y=x**2
plt.plot(x,x,marker="o")
plt.plot(x,y,marker="X")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1,6)
y = x**2
plt.plot(x,x,marker="o",c ="red")
plt.plot(x,y,marker="X",c="green")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1,6)
y = x**2
z = x**3
plt.plot(x,x,marker="o",c="red")
plt.plot(x,y,marker="X",c="green")
plt.plot(x,z,marker="*",c="black")
plt.show()
```
