```python

#! How to plot Multiple Plots (Lines) in same axes with legend
from matplotlib import pyplot as plt 
import numpy as np
x = np.arange(1,6)
y = x**2
z = x**3
plt.plot(x,x,marker="o",c = "red")
plt.plot(x,y,marker="X",c="green")
plt.plot(x,z,marker="*",c="black")
plt.show()
```
